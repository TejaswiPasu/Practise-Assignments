```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
p=0.6 
data=bernoulli.rvs(p,size=1000)
plt.bar([0,1],[sum(data==0),sum(data==1)],color=['red','green'],tick_label=['Failure(0)','Success(1)'])
plt.xlabel('outcome')
plt.ylabel('frequency')
plt.title('bernoulli')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters
mu = 70  # Mean
sigma = 10  # Standard deviation

# Generate Normal distribution data
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)  # Range of scores
pdf = norm.pdf(x, mu, sigma)  # Compute PDF

# Plot the Normal Distribution
plt.plot(x, pdf, color="red", label=f"N({mu}, {sigma})")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (={mu}, ={sigma})")
plt.legend()
plt.grid()
plt.show()
```
